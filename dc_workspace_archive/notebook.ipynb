## Secondly, visualise the companies by segregating them according to Market Cap in three categories: Large Cap(>20,000 crore), Mid Cap (5,000 to 20,000 crore) & Small Cap(<5,000).

```python
# Visualize Large market cap
large_cap.plot(x="Name", y="Market Cap", figsize=(20, 10))
plt.xticks(rotation=90)
plt.title("Companies with Large Market Cap greater than 20000")

# Visualize medium market cap
medium_cap.plot(x="Name", y="Market Cap", figsize=(20, 10))
plt.xticks(rotation=90)
plt.title("Companies with Medium Market Cap from 5000 to 200000")

# Visualize low market cap
low_cap.plot(x="Name", y="Market Cap", figsize=(20, 10))
plt.xticks(rotation=90)
plt.title("Companies with Low Market Cap less than 5000")

plt.show()
```

# Task 3:
## Thirdly, pick 10 random companies from the entire group and visualise the following through a line chart Find the Intrinsic Value of the company based on 3 cases of growth (g): Assume 3 Cases for g (Growth) : Good (15% Growth) ; Bad (-5% Growth); Best (25%Growth).

### Visualise these 10 companies on a line chart for all 3 cases of growth going forward against its current market cap.

# V = (EPS * (8.5 + 2g) * 6) / Y

**V** : Intrinsic Value

**EPS** : The Companys last 12 month earnings per share

**8.5** : The constant represents the appropriate P-E ratio for a no-growth company as proposed by Graham.

**g** : The company's long-term (five years) earnings growth estimate

**6** : The average Return of FDs (6%)

**Y** : The current yield on AAA corporate bonds

```python
# The Companys last 12 month earnings per share
EPS = stock_data["Earnings Per Share"]

# current AAA corporate bond yield in india is 7.282 %
Y = 7.282

# The company's long-term (five years) earnings growth estimate
g = stock_data["5Y CAGR"]
```

```python
# Intrinsic Value of 10 random companies
intrinsic_value = ((EPS * (8.5 + 2 * g) * 6) / Y) /100 
stock_data['Intrinsic Value %'] = intrinsic_value
stock_data.head(20)
```

```python
rand_comp = stock_data[["Name", "Intrinsic Value %"]].sample(n=10)
rand_comp
```

```python
rand_comp.plot(x="Name", y="Intrinsic Value %", kind='line', rot=90, figsize=(30, 10))
```

```python
good_growth = rand_comp[(rand_comp["Intrinsic Value %"] >= 15) & (rand_comp["Intrinsic Value %"] < 25)]
bad_growth = rand_comp[(rand_comp["Intrinsic Value %"] >= -5) & (rand_comp["Intrinsic Value %"] < 15)]
best_growth = rand_comp[rand_comp["Intrinsic Value %"] >= 25]
```

```python
good_growth.plot(x="Name", rot=90, figsize=(30, 10))
bad_growth.plot(x="Name", rot=90, figsize=(30, 10))
best_growth.plot(x="Name", rot=90, figsize=(30, 10))
plt.legend()
plt.show()
```
